```python
from pathlib import Path

import numpy as np
import tensorflow.keras
from PIL import Image, ImageOps

# Disable scientific notation for clearer prediction output.
np.set_printoptions(suppress=True)

MODEL_PATH = Path("keras_model.h5")
IMAGE_PATH = Path("test_photo.jpg")

if not MODEL_PATH.exists():
    raise FileNotFoundError("Missing keras_model.h5. Export the model from Teachable Machine and place it next to this notebook.")

if not IMAGE_PATH.exists():
    raise FileNotFoundError("Missing test_photo.jpg. Add the image you want to classify next to this notebook.")

model = tensorflow.keras.models.load_model(MODEL_PATH)

# Teachable Machine image models expect one 224x224 RGB image.
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

image = Image.open(IMAGE_PATH).convert("RGB")
resample_filter = Image.Resampling.LANCZOS if hasattr(Image, "Resampling") else Image.ANTIALIAS
image = ImageOps.fit(image, (224, 224), resample_filter)

image_array = np.asarray(image)
normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
data[0] = normalized_image_array

prediction = model.predict(data)
print(prediction)
```
